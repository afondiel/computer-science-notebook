```python
import sys
# sys.path.append('/content/aimet')  # After cloning repo
import torch
from torchvision.models import resnet18
from torchvision.models import mobilenet_v2

from aimet_common.defs import CostMetric, CompressionScheme
from aimet_torch.compress import ModelCompressor
from aimet_torch.defs import SpatialSvdParameters, ChannelPruningParameters
from Examples.torch.utils import pipeline
```

```python

# 1. Basic Model Setup
model = resnet18(pretrained=True).eval()
input_shape = (1, 3, 224, 224)

# 2. Compression Configuration
compression_tech = CompressionScheme.spatial_svd  # Options below

# 3. Execute Compression (Spatial SVD example)
spatial_svd_params = SpatialSvdParameters(
    mode=SpatialSvdParameters.Mode.auto,
    params=SpatialSvdParameters.AutoModeParams(
        GreedySelectionParameters(target_comp_ratio=Decimal(0.8))
    )
)

compressed_model, stats = ModelCompressor.compress_model(
    model=model,
    eval_callback=lambda model, _: 0.75,  # Dummy evaluation
    eval_iterations=1,
    input_shape=input_shape,
    compress_scheme=compression_tech,
    cost_metric=CostMetric.mac,
    parameters=spatial_svd_params
)

print(f"Compression stats: {stats}")
```
